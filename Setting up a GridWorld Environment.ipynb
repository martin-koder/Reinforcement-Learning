```python
env=Quarry(10)
env.reset()
env.show_env()
```

```python
env.resource_target # shows target amount of resource value to meet termination condition (10*N//4 where each resource is worth 10)
# here in practice the agent needs to collect three blocks of resource
```

```python
print('Take step out of the base  - define appropriate step action (up, down, left right), then call env.show_env() again')
env.step(str(input()))   # step returns the state index, reward for that state and the termination status
```
